```python
# A1 Bokeh is a Python library that provides interactive data visualization with HTML and JavaScript. 
# To create a Boken Plot we need to import bokeh.plotting module
from bokeh.plotting import figure, show, output_file
import bokeh.io
bokeh.io.output_notebook()
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv") #loading iris dataset into a dataframe
df
```

```python
bfig=figure(title="Iris Dataset", x_axis_label="Sepal Width", y_axis_label="Sepal Length") #loading iris dataset into a dataframe
bfig.circle(x=df['sepal_width'],y=df['sepal_length'])#loading iris dataset into the figure object
show(bfig)
```

```python
##Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

from bokeh.plotting import figure, show, output_file
import bokeh.io
bokeh.io.output_notebook()
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv") #loading iris dataset into a dataframe
df
```

```python
from bokeh.models import Legend
from pandas.core.arrays.interval import le
bfig=figure(title="Iris Dataset", x_axis_label="Petal Width", y_axis_label="Petal Length", background_fill_color="grey", outline_line_color="white") #creating a figure object and customising colors
bfig.line(x=df['petal_width'],y=df['petal_length'], line_color="red",legend_label="Petal points")#loading iris dataset into the figure object and adding a line glyph in bokeh
show(bfig)
```
