```python
#Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution
#of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to
#these features to improve normality?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
df=pd.read_csv('winequality-red.csv')
df.head()
```

```python
# PLot density plot for all features
plt.style.use('seaborn')
df.hist(bins=50, figsize=(20,15), ec = 'b')
plt.show()
```

Based on above Volatile Acidity,chlorides, free sulfur dioxide, total sulfur dioxide are exihibiting non-normaility 

```python
df['volatile acidity'] = df['volatile acidity'].apply(lambda x: np.log(x))
df['chlorides'] = df['chlorides'].apply(lambda x: np.sqrt(x))
df['total sulfur dioxide'] = df['total sulfur dioxide'].apply(lambda x: np.log(x))
df['free sulfur dioxide'] = df['free sulfur dioxide'].apply(lambda x: np.sqrt(x))
df['residual sugar'] = df['residual sugar'].apply(lambda x: np.sqrt(x))
```

```python
# PLot density plot for all features
plt.style.use('seaborn')
df.hist(bins=50, figsize=(20,15), ec = 'b')
plt.show()
```

Normality improved based on logarithimic transformations
