```python
##Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
##plot a scatter plot.Note: Also add title, xlabel, and ylabel to the plot.

##A2 A scatter plot is a diagram where each value in the data set is represented by a dot. 
##It is used to observe the relationship between two variables and how change in one affects the other. 


##In matplotlib, one can create a scatter plot using the pyplot’s scatter() function

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
##creating the scatter plot

plt.scatter(x,y)
##Adding chart labels and titles
plt.title("The Plot of X vs Y Values")
plt.xlabel("X-Axis ------------->")
plt.ylabel("Y-Axis ------------>")
```

```python
### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
##The subplot() function is used to create a grid of subplots within a single figure. 
##It allows to plot multiple graphs on the same figure with different axes and scales.
##The subplot() function takes two or three arguments: nrows, ncols, and index. 
##The nrows and ncols arguments specify the number of rows and columns of the subplot grid. 
##The index argument specifies the position of the subplot within the grid
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure()
##create four subplots with different line plots
plt.subplot(2,2,1) # select first subplot
plt.plot(x,y1,'r-') # plot red line
plt.title('Plot A') # add title
plt.grid()

plt.subplot(2,2,2) # select second subplot
plt.plot(x,y2,'b-') # plot blue line
plt.title('Plot B') # add title
plt.grid()

plt.subplot(2,2,3) # select third subplot
plt.plot(x,y3,'g-') # plot green line
plt.title('Plot C') # add title
plt.grid()

plt.subplot(2,2,4) # select fourth subplot
plt.plot(x,y4,'m-') # plot magenta line
plt.title('Plot D') # add title
plt.grid()
plt.grid()

# adjust spacing between subplots
plt.tight_layout()
plt.grid()

# show the figure
plt.show()
```

```python
###Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### A bar plot is a graph that represents categorical data with rectangular bars with heights proportional to the values they represent
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

ax=plt.bar(company,profit, color='blue') # creating the blue bart chart
plt.title("Company Profits for the year 2023 in Billions USD") # creating the chart title
plt.xlabel("Company Names----------->")# creating X axis label
plt.ylabel("Profit --------------->")# creating Y axis label
for i in range(0,np.count_nonzero(company)):
    plt.bar_label(ax, label_type='center', color='white')# creating datalabels to indicate values
plt.grid()
plt.show()
```

```python

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#A box plot is a type of chart that shows how data is distributed using five numbers: minimum, first quartile, median, third quartile, and maximum
```

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]
 
fig = plt.figure(figsize =(5, 5))
 
# Creating axes instance
ax = fig.add_axes([0,1,1,1])
 
# Creating plot
bp = ax.boxplot(data)
plt.title("Box chart") # creating the chart title
plt.xlabel("Box Values")# creating Y axis label

# show plot
plt.show()
```
