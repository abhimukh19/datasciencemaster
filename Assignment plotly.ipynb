```python
#Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
#scatter plot for age and fare columns in the titanic dataset.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as gb
df=pd.DataFrame(sns.load_dataset('titanic')) # loading titanic dataset in dataframe
df
```

```python
pfig= gb.Figure()
pfig.add_trace(gb.Scatter(x=df['age'],y=df['fare'], mode='markers', showlegend=True)) # loading scatter plot on a data with plotly
pfig.update_layout(xaxis_title='age', yaxis_title='fare') # updating x and y axis labels
pfig.update_layout(paper_bgcolor="grey") # changing background color
```

```python
##Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as ex
df2=pd.DataFrame(sns.load_dataset('tips')) # loading tips dataset in dataframe
df2
```

```python
pfig1= ex.box(x=df2['day'],y=df2['total_bill']) # loading box plot on a data with plotly express
pfig1.update_layout(xaxis_title='day', yaxis_title='total_bill') # updating x and y axis labels
pfig1.show()
```

```python
#Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
#the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
#column with the color parameter.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as ex
df3=pd.DataFrame(sns.load_dataset('tips')) # loading tips dataset in dataframe
df3
```

```python
pfig3=ex.histogram(x=df3['sex'], y=df3['total_bill'], pattern_shape=df3['smoker'], color=df3['day']) # ploting an histogram for sex and total bill with smoker as a pattern shape and day as color paramete
pfig3.update_layout(title="Tips dataset",xaxis_title='sex', yaxis_title='total_bill') # updating x and y axis labels
pfig3
```

```python
#Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
#the color parameter.
#Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
#dimensions parameter.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as ex
df4=pd.DataFrame(sns.load_dataset('iris')) # loading iris dataset in dataframe
df4
```

```python
pfig4= ex.scatter_matrix(data_frame=df4, dimensions=[df4['sepal_length'],df4['sepal_width'],df4['petal_length'],df4['petal_width']],color=df4['species'])  # creating the scatter matrix plot as per requirements
pfig4.update_layout(width=800, height=800)  # resizing the plot to fit the x and y axis
```

```python
#Q5. What is Distplot? Using Plotly express, plot a distplot.

#A5. A distplot is a plot that shows how a single variable is distributed. 
#It can combine a histogram with a curve that shows the density or frequency of the values 
#A distplot can help understand the shape, spread and outliers of a dataset

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as ex
df3=pd.DataFrame(sns.load_dataset('tips')) # loading tips dataset in dataframe
df3

```

```python
pfig5=ex.histogram(x=df3['total_bill'], color=df3['sex'],marginal="box") # ploting an displot for total bill color parameter set of sex
pfig5.update_layout(title="Tips dataset",xaxis_title='total bill') # updating y axis labels

pfig5 #We can see how the distribution of total bill varies by sex and also see summary statistics such as median, quartiles and outliers.
```
