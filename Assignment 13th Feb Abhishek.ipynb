```python
##Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
### A3 Errors in ArithmeticErrors are for mathematical operations error like ZeroDivisionError, OverflowError and Floatingpointerror
## example in code for  zerodivision error

try:
    # initialize a list
    list1=[1,2,3,4,5]
    print("Execution Started")
    for i in list1:
        i=i/0
        print(i)

except ArithmeticError as Aerror:
    print("Exception Thrown is: ", Aerror)

finally:
    print("Execution Terminated")
```

```python
## example in code for Overflow error
try:
    # initialize a list
    list1=[10.0,2000,200000000,3000000,40000000,5000000]
    print("Execution Started")
    for i in list1:
        i=i**2000000000000
        print(i)
    raise ArithmeticError
except ArithmeticError as Aerror:
    print("Exception Thrown is: ", Aerror)

finally:
    print("Execution Terminated")
```
