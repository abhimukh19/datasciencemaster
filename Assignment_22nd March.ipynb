```python
#Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
#you have collected data on the amount of time students spend studying for an exam and their final exam
#scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
import pandas as pd

data = {
    'Study Time (hours)': [2.5, 3.0, 1.5, 4.0, 2.0, 3.5, 2.0, 1.0, 4.5, 3.0, 2.5, 1.5, 4.0, 2.0, 3.5, 2.0, 1.0, 4.5, 3.0, 2.5, 1.5, 4.0, 2.0, 3.5, 2.0],
    'Final Exam Score': [80, 85, 60, 90, 75, 85, 70, 50, 95, 80, 75, 65, 90, 70, 85, 75, 55, 95, 80, 75, 65, 90, 70, 85, 75]
}

df=pd.DataFrame(data)
df
```

```python
df1=df.corr(method='pearson')
df1
```

```python
#There is a strong positive co-relation of 0.961 between study time and final score
```

```python
#Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
#Suppose you have collected data on the amount of sleep individuals get each night and their overall job
#satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
#variables and interpret the result.

import pandas as pd

data = {'Amount of Sleep': [7, 8, 6, 5, 9, 7, 8, 6, 5, 9, 7, 8, 6, 5, 9],
        'Job Satisfaction': [8, 9, 7, 6, 10, 8, 9, 7, 6, 10, 8, 9, 7, 6, 10]}

df = pd.DataFrame(data)
df
```

```python
df2=df.corr(method='spearman')
df2
```

```python
#The value of 1 indicates the there is strong relationship between Amount of Sleep and Job Satification
```

```python
#Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
#exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
#for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
#between these two variables and compare the results.

import pandas as pd
import random
Hours_of_exercise_per_week=[random.randint(5,30) for _ in range(50)]
BMI= [random.randint(21,26) for _ in range(50)]
data = {'hours_of_exercise_per_week': Hours_of_exercise_per_week,
        'BMI':BMI }

df = pd.DataFrame(data)
df
```

```python
df2=df.corr(method='pearson')
df2
```

```python
df3=df.corr(method='spearman')
df3
```

```python
#Getting negative correlation in pearson while positive in spearman showing difference of impact of excercise on BMI
```

```python
#Q4. A researcher is interested in examining the relationship between the number of hours individuals
#spend watching television per day and their level of physical activity. The researcher collected data on
#both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
#these two variables
import random
import pandas as pd

# Create a list of hours of TV per day
hours_of_tv = [random.randint(1, 5) for _ in range(50)]

# Create a list of levels of physical activity
levels_of_activity = [0, 1,2]
activity = [random.choice(levels_of_activity) for _ in range(50)]

# Create a dictionary with the data
data = {'Hours of TV per day': hours_of_tv,
        'Level of physical activity': activity}

df=pd.DataFrame(data)
df
```

```python
# Calculate the Pearson correlation coefficient for the two columns
dfoutput=df.corr()
# Print the correlation coefficient
dfoutput
```

```python
#There is a negative correlation between level of activity and watching TV
```

```python
#Q5. A survey was conducted to examine the relationship between age and preference for a particular
#brand of soft drink. The survey results are shown below:

import pandas as pd

data={'Age (Years)':[25,42,37,19,31,28],'SoftDrink Preference':['Coke', 'Pepsi','MountainDew','Coke','Pepsi','Coke']}
df=pd.DataFrame(data)
df
```

```python
df = pd.get_dummies(df, columns=['SoftDrink Preference'])
df
```

```python
dfoutput=df.corr(method='spearman')
dfoutput
```

```python
#There is positive correlation between age and Soft Drink Preference of Mountain Dew and Pepsi while Negative for Coke
```

```python
#Q6. A company is interested in examining the relationship between the number of sales calls made per day
#and the number of sales made per week. The company collected data on both variables from a sample of
#30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

import random
import pandas as pd

sales_calls_perday = [random.randint(10, 100) for _ in range(30)]

sales_made_per_week =[random.randint(100,1000) for _ in range(30)]

# Create a dictionary with the data
data = {'Sales_calls_perday': sales_calls_perday,
        'Sales_made_per_week': sales_made_per_week}

df=pd.DataFrame(data)
df
```

```python
# Calculate the Pearson correlation coefficient for the two columns
dfoutput=df.corr()
# Print the correlation coefficient
dfoutput
```

```python
#There is a positive correlation between calls and sales made
```
