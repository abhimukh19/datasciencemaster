```python
##Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
##Note: timepoint, signal, event, and region are columns in the fmri dataset.
##importing seaborn library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame(sns.load_dataset('fmri')) # loading fmri dataset in dataframe
df
```

```python
sns.lineplot(x='timepoint', data=df,y='signal', hue='region') # creating a line plot with x axis as timepoint and y axis as signal
```

```python
sns.lineplot(x='timepoint', data=df,y='signal', hue='event') # creating a line plot with x axis as timepoint and y axis as signal
```

```python
# Create a figure with 2 subplots arranged in one row and two columns
fig, axes = plt.subplots(1, 2,  figsize=(10, 6))
fig.suptitle("FMRI DATA SET")
sns.lineplot(x=df['timepoint'],y=df['signal'],  hue=df['event'], data=df['event'], ax=axes[0])  # creating a event sub line plot with x axis as timepoint and y axis as signal
sns.lineplot(x=df['timepoint'],y=df['signal'],  hue=df['region'], data=df['region'], ax=axes[1])# creating a region sub line plot with x axis as timepoint and y axis as signal
```

```python
##Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
##Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df1=pd.DataFrame(sns.load_dataset('titanic')) # loading titanic dataset in dataframe
df1
```

```python
# Create a figure with 2 subplots arranged in one row and two columns
fig, axes = plt.subplots(1, 2,  figsize=(10, 6))
fig.suptitle("Titanic SET")
sns.boxplot(x=df1['pclass'], y=df1['age'], data=df1, ax=axes[0])  # creating a box plot with x axis as pclass  and y as age
sns.boxplot(x=df1['pclass'], y=df1['fare'], data=df1, ax=axes[1])  # creating a box plot with x axis as pclass  and y as age
plt.tight_layout()
```

```python
##Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
##parameter for the 'cut' column of the diamonds dataset.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df2=pd.DataFrame(sns.load_dataset('diamonds')) # loading diamonds dataset in dataframe
df2
```

```python
##Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
flights=sns.load_dataset('flights') # loading flights dataset in dataframe
flights=flights.pivot("month", "year", "passengers")
```

```python
sns.heatmap(data=flights) # creating a heatmap with flights data as year as x axis and month as y axis and passengers as the heatmap scale
```
