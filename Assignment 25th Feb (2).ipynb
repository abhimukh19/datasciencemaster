```python
## Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

###The difference between loc and iloc is that loc selects rows and columns with specific labels from the index, while iloc selects rows and columns at particular positions in the index.
```

```python
##Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
#then find the output for both new_df.loc[2] and new_df.iloc[2].

# Import pandas library
import pandas as pd

course_name = ["Data Science", "Machine Learning", "Big Data", "Data Engineer"]
duration = [2,3,6,4]
df = pd.DataFrame(data = {"course_name" : course_name, "duration" : duration})
print("Original DataFrame")
print(df)
new_df= df.reindex([3,0,1,2])
print("\n New DataFrame")
print(new_df)

print("\n new_df.loc[2] output with index label as the identifier")
print(new_df.loc[2])

print("\n new_df.iloc[2] output with index position as the identifier")
print(new_df.iloc[2])

print("\n The difference is clear one is returning the value corresponding to an index label and other based on the index position")
```

```python
##Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
##calculates the difference between them in days, hours, and minutes using Pandas time delta. The
##program should prompt the user to enter the dates and display the result.

# Import pandas and datetime library
import pandas as pd
from datetime import datetime, timedelta, date
date1=pd.to_datetime(input("Please enter your from date in YYYY-MM-DD format"))
date2=pd.to_datetime(input("Please enter your to date in YYYY-MM-DD format"))
diff=pd.Timedelta(date2-date1)
print(f"The different between the two dates in days is:  {diff.days}, while in hours is : {((diff.days)*24)} and in minutes is : {((diff.days)*24*60)}")

```

```python
##Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
##column to a categorical data type. The program should prompt the user to enter the file path, column
##name, and category order, and then display the sorted data.
```

```python
# Import pandas library
import pandas as pd

# Prompt the user to enter the file path
file_path = input("Enter the file path: ")

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Prompt the user to enter the column name
new_column = input("Enter the column name: ")

# Prompt the user to enter the category order as a comma-separated list
category_order = input("Enter the category order: ")

# Convert the category order to a list of strings
category_order = category_order.split(",")

# Convert the column to a categorical data type with the specified order
df[new_column] = pd.Categorical(df[new_column], categories=category_order, ordered=True)

# Sort the data by the column and display it
df.sort_values(by=new_column, inplace=True)
print(df)
```

```python
##Q10 Write a Python program that reads a CSV file containing sales data for different products and
##visualizes the data using a stacked bar chart to show the sales of each product category over time. The
##program should prompt the user to enter the file path and display the chart.

# Import pandas library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta, date

# Create a sample DataFrame
df= pd.read_excel("SampleData.xlsx",sheet_name="SalesOrders")
# Changing the column names to make it readable
df.rename(columns = {'Item':'ProductCategory'}, inplace = True)
df.rename(columns = {'Total':'Sales'}, inplace = True)
# Changing the column to Datetype for Plotting and Pivoting
df['OrderDate']=pd.to_datetime(df['OrderDate'])
df['Year']=df['OrderDate'].dt.year
df
```

```python
# Creating a pivot table to sum the sales across years for each category of the product
df1=df.pivot_table(index='Year',columns='ProductCategory', values='Sales').replace(np. nan,0)

print(df1)
# Creating the stacked chart and adding data label for each stack in the stackchart
ax=df1.plot.bar(stacked=True, legend='upper right',title="Sales by Product Category over years")
for i in range(0,df['ProductCategory'].nunique()):
    plt.bar_label(ax.containers[i], label_type='center', color='white')
    i=i+1
plt.grid()
plt.show()
```
