```python
#Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you
#might choose one over the other.

#A1 Ordinal Encoding is a technique used to convert categorical data into numerical data by assigning each category a unique number. 
#Label Encoding is a similar technique, but it assigns each category a unique label.


#The main difference is that Ordinal Encoding preserves the order of the categories, while Label Encoding does not. 
#for example Ordinal Encoding can be used to categorize a dataset like [Bytes, Megabytes, Petabytes] where rank order matters 
#while label encoding can be used for [Rose, Tulip, Lotus] where numercial encoding is important and not necessarily rank order of the numerical codes

```

```python
#Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in
#a machine learning project.

#A2 It is a technique used to convert categorical data into numerical values based on the relationship between the categories and the target variable. 
#It is also known as ordered mean encoding

import pandas as pd

# Create a dictionary with age and income data
data = {'age': [25, 25, 35, 40, 40, 50, 55, 55, 65, 65],
        'income': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000]}

df=pd.DataFrame(data)
df
```

```python
mean_income=df.groupby('age')['income'].mean().to_dict()
mean_income
```

```python
df['income_encoded']=df['age'].map(mean_income)
df[['age','income_encoded']]
```

```python
#Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?

#A3 Covariance is used to determine the relationship between the movements of two random variables
#It is important as it is used to identify the direction of the relationship between two variables, whether it is positive or negative
#Covariance is calculated using the formula: Cov (X, Y) = Σ (Xi-µ) (Yj-v) / n, where Cov (X, Y) represents the covariance of variables X and Y
#Xi and Yj are the individual data points, µ and v are the means of X and Y, and n is the number of data points.
```

```python
#Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium,
#large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library.
#Show your code and explain the output.

import pandas as pd

# Create a dictionary for color, size and material
data = {'Color': ['red', 'green', 'blue', 'green', 'red'],
        'Size': ['small', 'medium', 'large', 'medium', 'small'],
        'Material': ['wood', 'metal', 'plastic', 'wood', 'metal']}
df=pd.DataFrame(data)
df
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
#creating a label encoder
le = preprocessing.LabelEncoder()
#applying on the columns
df['Color']=le.fit_transform(df['Color'])
df['Size']=le.fit_transform(df['Size'])
df['Material']=le.fit_transform(df['Material'])
df
```

```python
#Q7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity", and two
#categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/
#East/West). Calculate the covariance between each pair of variables and interpret the results.

import pandas as pd

# Create a dictionary for color, size and material
data =pd.read_csv('Tempdata.csv')
df=pd.DataFrame(data)
df
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

OneHotEncoder_var=preprocessing.OneHotEncoder()

# Fit and transform the "Weather Condition" column
df_encoded = OneHotEncoder_var.fit_transform(df[['Weather Condition']]).toarray()

# Replace the "Weather Condition" column with the encoded values
df['Weather Condition'] = df_encoded
df


```

```python
df['Wind Direction'] = pd.Categorical(df['Wind Direction'], categories=['North', 'South', 'East', 'West'], ordered=True)
df['Wind Direction'] = df['Wind Direction'].cat.codes
df
```

```python
import numpy as np

# Calculate covariance matrix
covariance_matrix = df.cov()

# Print the covariance matrix
print(covariance_matrix)
```
